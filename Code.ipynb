```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
```

```python
params = {
        'min_child_weight': [1, 5, 10],
        'gamma': [0.5, 1, 1.5, 2, 5],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6,0.8,1],
        'max_depth': [9,10,11],
        }
```

```python
xgb = XGBClassifier(learning_rate=0.01, n_estimators=600,objective='binary:logistic',booster='gbtree',random_state=100,
                    silent=True, nthread=1)
```

```python
folds = 5
param_comb = 5
random_search = RandomizedSearchCV(xgb, param_distributions=params, n_iter=param_comb,  n_jobs=4,  verbose=3, random_state=0 )
```

```python
random_search.fit(X_train, Y_train)
```

```python
print('\n Best estimator:')
print(random_search.best_estimator_)
```

```python
xgb_best = XGBClassifier(base_score=0.5, booster='gbtree', colsample_bylevel=1,
       colsample_bynode=1, colsample_bytree=0.6, gamma=5,
       learning_rate=0.02, max_delta_step=0, max_depth=11,
       min_child_weight=10, missing=None, n_estimators=600, n_jobs=1,
       nthread=1, objective='multi:softprob', random_state=1000,
       reg_alpha=0, reg_lambda=1, scale_pos_weight=1, seed=None,
       silent=True, subsample=1.0, verbosity=1)
```

```python
folds = 5
param_comb = 5
random_search_best = RandomizedSearchCV(xgb_best, param_distributions={'scale_pos_weight':[1,2,6,12]}, n_iter=param_comb,  n_jobs=4,  verbose=3, random_state=0 )
```

```python
random_search_best.fit(X_train, Y_train)
```

```python
print('\n Best normalized gini score for %d-fold search with %d parameter combinations:' % (folds, param_comb))
print(random_search_best.best_score_ * 2 - 1)
print('\n Best hyperparameters:')
print(random_search_best.best_params_)
```
